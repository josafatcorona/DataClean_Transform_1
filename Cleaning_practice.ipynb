```python
#Eliminamos columnas del dataset
fifa_players.drop(fifa_players.columns [[3,4,18]],axis = 1,inplace = True)
fifa_players.head(10)
```

```python
#validamos nulos
fifa_players.isnull().sum()
```

```python
fifa_players['Joined']
```

```python
#transformamos la columna Joined a date
#también almacenamos en columnas separadas los datos de día mes y año
x= j_d= j_m= j_y=[]
for item in fifa_players['Joined']:
#     x.append(datetime.strptime(item,"%b %d, %Y").date())
    j_date=datetime.strptime(item,"%b %d, %Y").date()
    x.append(j_date)
    j_d.append(j_date.day)
    j_m.append(j_date.month)
    j_y.append(j_date.year)
```

```python
fifa_players['Joined']=x
fifa_players['Joined_Day']=j_d
fifa_players['Joined_Month']=j_m
fifa_players['Joined_Year']=j_y
```

```python
#separamos la info de Team y Contract en diferentes columnas
team=[]
contract=[]
for element in fifa_players['Team & Contract']:
    team.append(list(filter(None,element.split("\n")))[0])
    contract.append(list(filter(None,element.split("\n")))[1])
```

```python
fifa_players['Team']=team
fifa_players['Contract']=contract
```

```python
fifa_players.groupby(fifa_players['Wage']).count()
```

```python
fifa_players['Wage']
```

```python
#Convertimos el salario a int(EUROS)
int_wage=[]
for wage in fifa_players['Wage']:
    wage=wage.replace('€','')
    wage=wage.replace('K','000')    
    int_wage.append(int(wage))
```

```python
fifa_players['Wage_E']=int_wage
fifa_players.describe()
```
