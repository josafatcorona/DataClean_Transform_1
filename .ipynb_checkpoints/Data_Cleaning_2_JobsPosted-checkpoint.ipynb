```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import collections
import re
```

```python
job_data=pd.read_csv('datasets/Uncleaned_DS_jobs.csv')
job_data.head()
```

```python
#let's explore Job Description
job_data['Job Description'].values[0]
```

```python
# buscamos si algunas de las herramientas más utilizadas son requeridas para el puesto
python=job_data['Job Description'].apply(lambda x: 'python' in x.lower())
excel=job_data['Job Description'].apply(lambda x: 'excel' in x.lower())
hadoop=job_data['Job Description'].apply(lambda x: 'hadoop' in x.lower())
excel=job_data['Job Description'].apply(lambda x: 'excel' in x.lower())
tableau=job_data['Job Description'].apply(lambda x: 'tableau' in x.lower())
big_data=job_data['Job Description'].apply(lambda x: 'big data' in x.lower())
power_bi=job_data['Job Description'].apply(lambda x: 'power bi' in x.lower())
senior=job_data['Job Description'].apply(lambda x: 'senior' in x.lower())
junior=job_data['Job Description'].apply(lambda x: 'junior' in x.lower())
```

```python
job_data['Python']=python
job_data['Excel']=excel
job_data['Hadoop']=hadoop
job_data['Tableau']=tableau
job_data['big_data']=big_data
job_data['power_bi']=power_bi
job_data['Senior']=senior
job_data['Junior']=junior
job_data.head()
```

```python
#retiramos  numeros del company name
company = job_data['Company Name'].apply(lambda x: re.sub(r"\n\d+(\.\d+)?", "", x))
company
```

```python
job_data['Company_Name']=company
job_data.head()
```

```python
#contamos las palabras más utilizadas en el job description
words={}
for job_d in job_data['Job Description']:
    for word in job_d.split():        
        if word in words.keys():            
            words[word.lower()]+=1
        else:
            words[word.lower()]=0
            
```

```python
# ordemanos las palabras que mas se repiten
sorted_words = sorted(words,key=words.get,reverse=True)
print(sorted_words)
```

```python
word_counts=collections.Counter()
for job_d in job_data['Job Description']:
    word_counts.update(job_d.lower().split())
```

```python
sorted_words_2=sorted(word_counts,key=word_counts.get,reverse=True)
sorted_words_2
```

```python
job_data['Salary Estimate'].value_counts()
```

```python
# Covertimos Salary a int, min max y avg
job_data['min_salary']=0
job_data['max_salary']=0
job_data['avg_salary']=0
for i in range(len(job_data)):
    try:
        job_data.loc[i,"min_salary"]=int(job_data['Salary Estimate'][i].split(" ")[0].split("-")[0].replace("$","").replace("K",""))
        job_data.loc[i,"max_salary"]=int(job_data['Salary Estimate'][i].split(" ")[0].split("-")[1].replace("$","").replace("K",""))
    except:
        job_data.loc[i,"min_salary"]=int(job_data['Salary Estimate'][i].split("(E")[0].split("-")[0].replace("$","").replace("K",""))
        job_data.loc[i,"max_salary"]=int(job_data['Salary Estimate'][i].split("(E")[0].split("-")[1].replace("$","").replace("K",""))
    finally:
        job_data.loc[i,"Salary Estimate"]=str(job_data.loc[i,"min_salary"])+"-"+str(job_data.loc[i,"max_salary"])
        job_data.loc[i,"avg_salary"]=np.mean([job_data.loc[i,"min_salary"],job_data.loc[i,"max_salary"]])
```

```python
#contamos por estado, el numero de empleos | remplazamos nombre largos por 2 letras
job_data['Job_State']=job_data['Location'].apply(lambda x: x.split(",")[-1].strip())
job_data['Job_State'].replace(['United States','Texas','California','New Jersey','Remote','Utah'],["US","TX","CA","NJ",np.nan,np.nan],inplace=True)
job_data['Job_State'].value_counts()
```
